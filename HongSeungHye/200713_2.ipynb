    - force 하는 이유: base 싱크가 안 맞을 떄 문제 발생. local로 github을 덮어쓰고 싶을 때 사용.
    - base 싱크가 맞고, 수정본이 없는, 6개 다 같은데 위에 더 있으면 제출 가능.

    - 팀플 maintaining -> force 잘 안 씀. 같이 쓰는 repo니까.
    - fork해서 새롭게 작업, pull request로 제출하는 상황 -> fork를 뜬 repo는 commit 내용 수정 위해 force 많이 사용 가능.
    - 기여 입장에서는 fork한 repo 중심으로 force 많이 사용 가능. then pull request 수정 효과.
    - maintainer 입장에서는 force 수정 잘 하지 않음.
    - 논리적으로는 같이 쓰게 되는 repo는 수정 않음. fork 뜬 repo는 수정 많이 가능.

    - 팀플을 기준으로 포크를 떠서 repo를 새로 만들었음. github 상에서 복사. 새로 복사하면 그 때 역사를 기준으로 고정. 원 repo가 바뀐다고 following 되는 건 아님.
    - clone -> 노트북에 그 history 전체를 받게 됨. 기본적으로 다른 프로젝트에 같이 작업 또는 협업 또는 오픈소스 참여 시 이 구조를 갖게 됨.
    - 왼쪽이 upstream, 오른쪽이 origin. origin만 있으면 거기만 push하고, push 권한으로 다른 사람과 작업 가능하겠지만 보통 이렇게 기본 중앙이 되는 repo가 있고, fork해서 만들고, 내 작업은 origin에 관리하고, 작업이 되면 pull request로 작업 가능.
    - pull request는 github의 기능; push를 먼저 해야 함. origin에 올리고, 버튼 눌러서 제출 가능.
    - 브랜치명 선택 - fork를 뜬 프로젝트인데, branch를 누르면 develop 브랜치가 있음. 그걸 보고 pull request를 누르면 차이점 보고 제출 가능.
    - pull request: github 기능. 내가 한 작업을 fork 뜬 데에 올리면 pull request로 제출 가능. 바로 merge되지는 않음. maintainer가 merge 눌러주면 merge됨. 지금은 pull request 제출까지만 진행.

- git 사용 이유와 장점

- 소스코드 변화과정 역사 저장 및 업로드, 제출 등의 시나리오를 진행.
- 흐름을 알고, 세부적인 건 트레이닝하면서 진행.

- branch
    - 같은 경로에 브랜치 만들고, 파일 만들고, commit해서 history 만들고 branch 변경 -> 폴더 내에 파일이 없어짐. 기존 브랜치로 가면 폴더에 파일이 다시 생김.
    - 같은 폴더인데, 전혀 다른 세상처럼 운용 가능.
    - 단, history에 담겨있어야 함.
    - branch가 다르면 같은 폴더라도 다른 폴더처럼 운용 가능.
        - -> 여러 파일이 생기고 변경되는 걸 전혀 다르게 운용 가능. 같은 폴더여도.
    - 필요 이유: 작업 중 작업 내용을 버리고 싶지도 않고, 압축파일로 만들면 용량이 커짐. 다른 폴더에 놓자니 시스템 설정, path 정보가 바뀜. 현 위치에 두면서 다른 작업도 하고 싶을 때.
    - 한 폴더임에도 다른 역사를 운용 가능.
    - 보통 작업 시, 내가 새로운 팀 프로젝트 작업 진행 -> develop이라는 브랜치 명칭보다는 작업하고자 하는 명칭으로 브랜치 만듬. config 관련 fix 작업 등을 하려면 fix-config로 만드는 등 하려는 작업에 맞춰서 브랜치를 만들도록 함. 작업 명칭을 쓰는 게 보기 좋음. 그냥 fix 1, 2, 3... -> 뭘 했는지 알 수 없음. so 브랜치 명칭은 키워드를 따서 해주는 게 좋음. 그리고 branch 따서 작업한 내용을 그대로 pull request로 제출 가능; 작업시 master 브랜치에서 따고, 작업한 내용을 제출하는 식으로 굴러감.
   

- rebase
    - fork, clone으로 local에 저장. 작업 시작하는 게 base. base에 새로운 작업 진행 가능.
    - base는 바뀜; upstream에 바뀜에 따라. then 모두 최신 base로 바꿔줘야 함. upstream이 기준이고, 변화에 따라 base 바꾸어줘야.
    - 최초에는 upstream = origin = local.
    - 여기서 pull request하려면 push 후 버튼 눌러서 제출. 그런데 마음대로 잘 안 됨. 오픈소스 개발자 협업하다보면 리눅스 커널 등도 6 ~ 7000명이 개발. 한 폴더 안의 소스 코드를. then 내가 작업한 걸 올리기 전에 다른 사람 들어갈 수 있음. so 내가 하기 전에 다른 커밋이 들어가 있을 수 있음. 현실적으로 이럴 수는 있음. 기다려라, ... 근데 6~7천명이 그럴 수는 없음. then 트러블 슈팅 어떻게 하나? 싱크 안 맞음. fork? clone? 또 새로 올라오면?
    - base만 교체하면 됨. 이런 상황이 있을 떄 maintainer: please rebase라고 답변. 내가 과거에 기준으로 commit을 만들어 제출 -> 싱크 안 맞아서 merge 안 됨 -> rebase 다시 해서 제출.
    - rebase 설명이 검색을 많이 해보면 merge와 비교, 트리... 여러 설명하지만, 이런 전제로 설명하는 게 맞을 것. 오픈소스 협업 및 오픈소스 메커니즘 사용하는 기업들이 close 소스라도 pull request 방식으로 진행. 이 떄 base 교체 필요 발생.
    - base 교체 시 petch 명령으로 최신 내용 가져와야 함. 이후 rebase 명령 -> base 교체 작업 이루어짐.
    - 처음 보는 입장에서는 어려울 수 있음. 그러나 명령어로는 간단. git fetch upstream master (업스트림에서 마스터 가져옴)
    - 가져온 브랜치가 upstream 브랜치에서 master 브랜치를 가져오면 로컬에 upstream/master라는 브랜치가 생김. then 거기 맞춰서 rebase하겠다는 의미.
    - 이 2가지 명령으로 종료. rebase 과정, 이유를 몰라서 설명이 필요할 뿐, 실제 사용시 2초. rebase 과정, 요청 과정이 복잡해보이나 이해하고 작업하면 어렵지 않음.
    - pull = petch + merge. 그냥 petch하게 되면 .git에 저장하게 됨. 안전히 가져오고, rebase로 하면 3가지 과정 거침.
        1. 되감기를 먼저 진행.
        2. base만 남았으니 base를 교체.
        3. 감았던 걸 다시 올림.
    - 이상이 rebase 동작 과정.
    - 아까 문제: rebase 명령 + enter -> 되감기를 먼저 진행. 3step 진행. 되감기, base 교체, 다시 올리기. 첫 스텝 = rewind = 과거로 복구. 현재 경로 = 최신 히스토리 내의 폴더. 되감기 -> 과거로 돌아감 -> 지금 만든 폴더는 과거에는 없었음. rewind하면 과거로 되돌리게 됨. base 교체 위해서. 그 과정에서 shell이 갖고 있는 현재 경로가 최근에 만든 폴더라서 경로 문제 발생. 과거에는 존재하지 않는 폴더가 현재 경로가 되니까.
    - 되감기 후 수정. 올리다가 문제 발생 가능할까? 올릴 떄 아래가 수정된 상황이라면 올리다 문제 발생 가능. but rewind는 내 수정 기록 기준 되감기. 문제 생기면 안 되는 상황. 내가 수정한 건 다 되감은 상태. 충돌이 날 수가 없음. 한 게 없는데.
    - 올렸음. 그럼 그 다음 commit 얹기는 충돌 가능할까? 가능. 4, 5가 내가 건드린 파일을 건드렸을 경우 complete가 날 수 있음.
    - 해결법은: 수동 해결 필요. 지워서 남길 거 남기고 작업해서 올리면 됨.
    - complete 나는 경우: maintainer가 force로 history 수정한 경우, 변경사항 올리다가 문제 가능. commit 1, 2, 3 내용이 달라질 경우. 충돌 발생 가능. so maintainer force 잘 안 씀; history 관리 제대로 하기 위해 force 수정 시 6~7천명이 다 complete 발생. base 교체하다가. so 그런 문제로 인해 force로 수정하지 말라고 권고. 그래도 하는 경우 있음. 그럼 어쩔 수 없음. 그런 식으로 진행.
    - then force를 아예 안 쓰나? 그건 아님. 관리자는 함부로 고치면 안 되겠지만, origin은 많이 수정 가능. fork 뜬 request고, 제출할 commit은 upstream에 제출. 보통 pull request 해서 제출하면 discussion을 거치게 됨. then 그거에 맞춰 수정하고 다시 branch를 force로 갱신되면 pull request 갱신 가능. origin에서는 force 많이 사용 가능. contributor 입장에서는 fork 뜬 프로젝트 중심으로는 force 많이 사용 가능.
    - base 교체 컨셉만 가져가더라도 어느 정도 많이 가져가는 것.

- git 사용 이점.
    - 협업. hello world, 안드로이드 app에 git 필요 없지만 협업에서 큰 빛을 발휘.
    - 배경:
        - 변경성: SW 자체가 변경이 많이 일어나고.
        - 제조가 아니라 개발: 4명이서 개발하다가 20% 개발헀는데 속도 느리다고 사람 더 데려올 수 없음.
        - 비가시성은? SW의 특징. 보이지 않음. 소스코드는 보임. 소스코드 컴파일해서 만들어진 binary는 보려면 볼 수는 있음. 보이지 않는 건 실행중인 프로그램. 설계도면 = 소스코드. 구현된 것 = 건물. 건물에 뭐가 안 지어졌는지 구현체 보이지만, 소스코드로 만든 프로그램은 실제 실행이 구현체인데 눈에 보이지 않음. 이런 특징으로 인해 협업 어려움.
    - 어떻게 개발할 지 논의하며 만들어진 게 SW 공학. 그런 아이디어 중 하나가 commit 단위 개발. 쉽게 생각할 수 있지만 뛰어난 협업 방식. 현존 협업 방식 중 가장 훌륭. pull request로 작업 = commit 단위로 만들어 제출하고, commit 단위 test, review, discussion 작업 진행.
    - 한 폴더 아래 6천명 개발 가능한 방법. 과거에는 이런 방식 없어서 협업 힘들었으나 이제는 commit 단위 개발이 잘 이루어져 꼭 open하지 않더라도 commit 단위 방식으로 집단지성 발휘 가능.
    - 정리
        - 비가시성 극복 X. history는 가시화되나 메모리 상에 돌고 있는 프로그램은 보이지 않음. but commit 단위 관리 -> 소스 코드 line by line 관리 -> 버그 사전 차단 및 방지 가능. so 많은 기업들이 git을 통해 commit 단위 개발.
        - commit 단위 개발 시 코딩 스타일 존재. if문 쓸 떄 중괄호 어디에 넣을 지, else는 어디 적을 지... 정답 없음. 큰 문제. 6천명 공동 개발 -> 코드 수정 이유: 중괄호 조정 이런 경우 개발 못 함. 코딩 스타일 충돌. 작업 시 파이썬을 쓴다면 django 프로젝트 등의 파이썬 코드 스타일 정의되어 있음. copy해서, 어떻게 코드 스타일을 쓰고 있는지, C++을 한다면 구글 코드 스타일... 그런 걸 합해서 진행하게 되면 깔끔히 정리 가능. 코드 스타일 중요 가능.
        - line by line 관리 -> 추적 용이.
- commit 단위 개발 시 history 존재. commit message 긴 경우 많고 이상적. 왜 수정했는지에 대한 정보 수정 가능. 보통 한 줄만 적는 경우도 있지만 그건 title이고, 수정한 이유 꼼꼼히 적는 것 중요. 협업 중 보면 10라인 중 8번쨰 line 누가 만들었는지 commit 통해 추출 가능. 소스코드 천 라인 중 어떤 문제가 있어서, 6번쨰 line에 문제가 있어서, 왜 나왔는지 history 관리 위해 봤는데 update 등만 쓰여 있으면 헙업 불가. 수정 이유와 방식이 정리되어 있지 않으면 문제 있는 소스 코드에 대해 커밋 히스토리 추출해서 fix 불가. 통째로 봐야 하는 상황 발생. commit 단위로 하나의 목적에 맞춰서, 여러 함수 만들듯 read, discussion하기 좋은 단위로 쪼개 관리해야.
- commit은 빨강, 초록, 추가, 빠진 것 등.
- pull request 보내면 discussion 가능. 한 줄 떄문에 긴 discussion도 가능. 당장 런칭해야 하는데. pull request하는 컨셉 자체가 명확. 오픈소스 개발 방식은 라이브러리, 프레임워크, os 등 의존도 높은 개발의 경우 적합. 빠른 개발 시 안 맞을 수 있음. 느리더라도 정확하게 가자는 것. 날이 갈 수록 갈아엎을 상황 많음. 라인 바이 라인 명확히 관리 시 그런 부분 해결 가능.
- git 사용 이유: 이런 방식 사용 시 코드 품질 높아짐. 오픈소스 역시 안전, 단단, 강하니까 사용. 좋을 수 있는 이유: 대표적으로 commit 단위 개발도 뺴놓을 수 없음. commit 단위 개발로 품질이 오르고, 다른 회사도 오픈소스 운영 않아도 pull request 방식으로 개발하게 됨.

- 오픈소스 개발 중 frontend 개발 하고 싶다 -> 오픈소스 안의 commit 단위가 교육적으로도 좋음. 유명한 FB, 구글, 애플 개발자가 왜 코드를 수정했고, 어떻게 수정했는지에 대한 discussion까지 전부 볼 수 있음. 그런 내용들도 가져갈 수 있고, 소스코드 reading 시에도 좋은 접근 가능. 단순 main 함수뿐 아니라 commit 단위 소스 코드 reading -> 원하는 단위 읽어들이기 가능. 이런 효과 존재.
